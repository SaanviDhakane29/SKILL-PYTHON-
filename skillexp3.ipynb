```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/autompg.csv')
print(df.head())
```

```python
#scatterplot
sns.scatterplot(data=df, x='horsepower', y='mpg', hue='origin')
plt.title('Scatter Plot: MPG vs Horsepower by Origin')
plt.show()
```

```python
sns.histplot(df['mpg'], kde=True)
plt.title('Histogram of MPG with KDE')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.barplot(data=df, x='origin', y='mpg')
plt.title('Bar Plot: Mean MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('Mean MPG')
plt.show()
```

```python
sns.lineplot(data=df, x='model year', y='mpg', marker='o')
plt.title('Line Chart: MPG over Model Year')
plt.xlabel('Model Year')
plt.ylabel('MPG')
plt.show()
```

```python
sns.boxplot(data=df, x='origin', y='mpg')
plt.title('Box Plot: MPG Distribution by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
df = pd.read_csv('/content/drive/MyDrive/autompg.csv')
print(df.head())
df['mpg_zscore'] = zscore(df['mpg'])
outlier_threshold = 3
outliers = df[df['mpg_zscore'].abs() > outlier_threshold]
print("Outliers based on Z-score:")
print(outliers[['mpg', 'mpg_zscore']])
```
